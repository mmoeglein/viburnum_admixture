```python
#this notebook refines plots for previous strcuture runs
#starting with structure runs from 2021_denovo_structure
#load packages 
import ipyrad as ip
import ipyrad.analysis as ipa
import pandas as pd
import toyplot
```

```python
#load path to the .snps.hdf5 database file for each assembly (parents, parents+hybrids, and just hybrids)
data_p = "/home/mm3465/scratch60/ipyrad/ipyrad_denovo_parent/parent_outfiles/parent.snps.hdf5"
data_ph = "/home/mm3465/scratch60/ipyrad/ipyrad_denovo_parent_hybrid/parent_hybrid_outfiles/parent_hybrid.snps.hdf5"
data_h = "/home/mm3465/scratch60/ipyrad/ipyrad_denovo_hybrid/hybrids_outfiles/hybrids.snps.hdf5"
data_hartout = "/home/mm3465/scratch60/ipyrad/ipyrad_denovo_hartout/hartout_outfiles/hartout.snps.hdf5"
data_denovo_yas = "/home/mm3465/scratch60/ipyrad/ipyrad_denovo_yas/denovo_yas_outfiles/denovo_yas.snps.hdf5"
data_denovo_zin= "/home/mm3465/scratch60/ipyrad/ipyrad_denovo_zin/denovo_zin_outfiles/denovo_zin.snps.hdf5"
```

```python
#reload structure object for denovo hybrids only
struct = ipa.structure(
    name="denovo_hybrid_k23_mincov3_5",
    data=data_h,
    mincov=0.3,)
```

```python
#hybrids: k=2
#lautum on left/mauve/#BC8F8F, jucundum on right/teal/#02818a
k = 2
table = struct.get_clumpp_table(k)
table = table.sort_values(by=[0])
#colors = ["#02818a", "#d0d1e6", "#BC8F8F", "#67a9cf", "#014636"]
colors = ["#02818a","#BC8F8F"]
# build barplot
canvas = toyplot.Canvas(width=3000, height=1000)
axes = canvas.cartesian(bounds=("10%", "90%", "10%", "45%"))
axes.bars(table, color=colors)
# add labels to x-axis
ticklabels = [i for i in table.index.tolist()]
axes.x.ticks.locator = toyplot.locator.Explicit(labels=ticklabels)
axes.x.ticks.labels.angle = -60
axes.x.ticks.show = True
axes.x.ticks.labels.offset = 10
axes.x.ticks.labels.style = {"font-size": "18px"}
```

```python
#same as above but changing the colors
#hybrids: k=2
#lautum on left/yellow, jucundum on right/pink
#potential yellows: "#EEB422"
k = 2
table = struct.get_clumpp_table(k)
table = table.sort_values(by=[0])
colors = ["#C67171", "#CD8500"]
# build barplot
canvas = toyplot.Canvas(width=3000, height=1000)
axes = canvas.cartesian(bounds=("10%", "90%", "10%", "45%"))
axes.bars(table, color=colors)
# add labels to x-axis
ticklabels = [i for i in table.index.tolist()]
axes.x.ticks.locator = toyplot.locator.Explicit(labels=ticklabels)
axes.x.ticks.labels.angle = -60
axes.x.ticks.show = True
axes.x.ticks.labels.offset = 10
axes.x.ticks.labels.style = {"font-size": "18px"}
```
